```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
points = 10000
```

```python
def plot_3_dimensions(A, title, x):
    b = np.matmul(A, x) # this is the ellipsoid
    ax = plt.axes(projection ='3d')
    ax.set_title(title)
    ax.scatter(b[0], b[1], b[2])
    plt.show()
```

```python
def plot_2_dimensions(A, title, x):
    b = np.matmul(A, x) # this is the ellipsoid
    plt.plot(b[0], b[1], 'o')
    plt.title(title)
    plt.show()
```

```python
def cond(A):
    return np.linalg.cond(A)

def determinant(A):
    return np.linalg.det(A)
```

```python
# Unit Circle
x = np.ndarray((2,points),dtype = np.float64)

pi = np.pi

for i in range(points):
    angle = random.uniform(0,2*pi)
    x[0][i] = np.cos(angle)
    x[1][i] = np.sin(angle)

print("x-coordinates:")
print(x[0])
print("y-coordinates:")
print(x[1])

plt.plot(x[0], x[1], 'o')
plt.title('Unit Circle')
plt.show()
```

```python
# a

# Rectangle Matrix (m != n)
A = np.array([
    [-1/np.sqrt(2),0],
    [0,-1/np.sqrt(2)],
    [-1,1]
])
plot_3_dimensions(A, "Part a", x)

print("Condition Number of matrix: ", cond(A))
```

```python
# b

# Rectangle Matrix (m != n)
B = np.array([
    [-2,1,2],
    [0,2,0]
])

x = np.vstack((x,np.ones((1,int(points)))))

plot_2_dimensions(B, "Part b", x)
x = np.delete(x,2,axis = 0)

print("Condition Number of matrix: ", cond(B))
```

```python
# c

# Square Matrix (m = n)
C = np.array([
    [1,0.9],
    [0.9,0.8]
])
plot_2_dimensions(C, "Part c", x)

print("Condition Number of matrix: ", cond(C))
print("Determinant of matrix: ", determinant(C))
if (determinant(C) == 0):
    print("Non Invertible Matrix")
else:
    print("Invertible Matrix")
```

```python
# d

# Square Matrix (m = n)
D = np.array([
    [1,0],
    [0,-10]
])
plot_2_dimensions(D, "Part d", x)

print("Condition Number of matrix: ", cond(D))
print("Determinant of matrix: ", determinant(D))
if (determinant(D) == 0):
    print("Non Invertible Matrix")
else:
    print("Invertible Matrix")
```

```python
# e

epsilon = [10, 5, 1, 10**-1, 10**-2, 10**-4, 0]
for e in epsilon:
    # Square Matrix (m = n)
    E = np.array([
        [1,1],
        [1,e]
    ])
    plot_2_dimensions(E,"Part e (epsilon = "+str(e) + ")" , x)
    
    print("Condition Number of matrix: ", cond(E))
    print("Determinant of matrix: ", determinant(E))
    if (determinant(E) == 0):
        print("Non Invertible Matrix")
    else:
        print("Invertible Matrix")
```

```python
print("Yes there is a relation between the condition number and the determinant. When the determinant is 0, i.e, the matrix is non-invertible, then the condition number is infinite. Otherwise it is finite")
```
